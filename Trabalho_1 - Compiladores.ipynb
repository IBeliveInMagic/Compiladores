```python
!# Função para saber em qual aresta o caracter deve prosseguir

def pertence(string):
    if string in alfabeto:
        if string in D:
            return 'D'
        elif string in L:
            if string == 'e':
                return 'e'
            elif string == 'E':
                return 'E'
            return 'L'
        else:
            return string
    else:
        return 'nulo'
```

```python
# Estados de aceitação
aceitacao = [1,3,6,7,9,10,11,12,13,14,16,17,18,19,20,21]
```

```python
def call_Token(string):
    if string == 1 or string == 3 or string == 6:
        return 'NUM'
    elif string == 7:
        return 'id'
    elif string == 9:
        return 'Lit'
    elif string == 10 or string == 11 or string == 12:
        return 'Opr'
    elif string == 13:
        return 'Opm'
    elif string == 14:
        return 'Rcb'
    elif string == 16:
        return 'Comentário'
    elif string == 17:
        return 'Fc_p'
    elif string == 18:
        return 'Ab_p'
    elif string == 19:
        return 'Pt_v'
    elif string == 20:
        return 'Vir'
    elif string == 21:
        return 'EOF'
    else : 
        return 'ERRO'
```

```python
def TESTE(atual,buffer):
    tupla['classe'] = call_Token(atual)
    tupla['tipo'] = 'Nulo'

    if tupla['classe'] == 'id':

        if (buffer,buffer) in tabela_simbolos:
            tupla['classe'] = buffer
            tupla['tipo'] = buffer
        else:
            tabela_simbolos['id',buffer] = 'Nulo'

    if tupla['classe'] == 'NUM':
        if atual == 1:
            tupla['tipo'] = 'inteiro'
        else:
            tupla['tipo'] = 'real'

    tupla['lexema'] = buffer
    return tupla['classe']
```

```python
# Abra o arquivo para leitura
with open('ProgramaFonte.txt', 'r') as arquivo:

    # Itere pelas linhas do arquivo
    numero_linha = 1
    linha = arquivo.readline()
    while linha:

        buffer = ""
        atual = 0
        # Itere pelos caracteres da linha
        numero_coluna = 1
        indice = 0
        tamanho_linha = len(linha)
        
        while indice < tamanho_linha:
            caractere = linha[indice]
            # percorrer os estados (DFA)------------------------
            c = pertence(caractere)
            #print(atual , caractere , c)

            if (atual, c) in arestas :
                
                next_states = arestas[(atual, c)]
                atual = next_states[0]
                buffer += caractere
                indice += 1
                numero_coluna += 1

            else:
                if TESTE(atual,buffer) != 'ERRO':
                    print(f"Classe: {tupla['classe']}, Lexema: {tupla['lexema']}, Tipo: {tupla['tipo']}")
                else:
                    if caractere != ' ' and caractere != '\n' and caractere != '\t':
                        tupla['classe'] = call_Token(caractere)
                        tupla['tipo'] = 'Nulo'
                        tupla['lexema'] = caractere
                        print(f"Classe: {tupla['classe']}, Lexema: {tupla['lexema']}, Tipo: {tupla['tipo']}")
                        print(f"ERRO léxico – Caractere inválido na linguagem . Linha: {numero_linha} , coluna: {numero_coluna}")
                    indice += 1
                    numero_coluna += 1

                if atual != 8 and atual != 15:
                    atual = 0
                    buffer = ""
                else:
                    buffer += caractere
        linha = arquivo.readline()
        numero_linha += 1
    if not linha:
        print(f"Classe: {tupla['classe']}, Lexema: {tupla['lexema']}, Tipo: {tupla['tipo']}")
        
        tupla['classe'] = call_Token(21)
        tupla['tipo'] = 'Nulo'
        tupla['lexema'] = 'EOF'
        print(f"Classe: {tupla['classe']}, Lexema: {tupla['lexema']}, Tipo: {tupla['tipo']}")
```
